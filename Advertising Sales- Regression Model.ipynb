```python
# Import libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
```

```python
# Read csv data
df = pd.read_csv('https://raw.githubusercontent.com/PriyaVj-UW/DS-Playground/master/Advertising.csv')
```

#### Linear Regression Model

```python
linear_model = LinearRegression()

def linearRegression(col1,col2,Title):
    x = df[col1].values.reshape(-1,1)
    y = df[col2].values.reshape(-1,1)
    
    linear_model.fit(x,y)
    y_pred = linear_model.predict(x)
    
    #plot the output
    plt.scatter(x,y)
    plt.plot(x,y_pred,linewidth=5)
    plt.grid(True)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(Title)
    plt.show()
    
    print('Intercept: {0:.5f}'.format(linear_model.intercept_[0]))
    print('slope: {0:.5f}'.format(linear_model.coef_[0][0]))

linearRegression('TV','sales','TV vs Sales')
linearRegression('radio','sales','Radip vs Sales')
linearRegression ('newspaper','sales','Newspaper vs sales')
```
